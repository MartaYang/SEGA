```python
import numpy as np
import os
from PIL import Image
```

```python
train_clsnames = "/data/FSLDatasets/tieredImageNet/train_classnames.txt"
TRAIN_TIERED_IMAGE_DATASET_DIR = "/data/tiered_imagenet/train"
val_clsnames = "/data/FSLDatasets/tieredImageNet/val_classnames.txt"
VAL_TIERED_IMAGE_DATASET_DIR = "/data/tiered_imagenet/val"
test_clsnames = "/data/FSLDatasets/tieredImageNet/test_classnames.txt"
TEST_TIERED_IMAGE_DATASET_DIR = "/data/tiered_imagenet/test"

all_classnames = []
```

```python
train_data = np.empty((448695,84,84,3), dtype='uint8') # 448695 is the total number of samples in training set.
train_label = np.empty(448695, dtype='uint16')
train_n = 0
with open(train_clsnames, "r") as f:
    for idx, line in enumerate(f):
        line = line.strip('\n')
        line = line.strip('\r')
        line = line.strip()
        print(idx, line)
        all_classnames.append(line)
        for file in os.listdir(os.path.join(TRAIN_TIERED_IMAGE_DATASET_DIR, line)):
            image = Image.open(os.path.join(TRAIN_TIERED_IMAGE_DATASET_DIR, line, file)).convert('RGB')
            img_ndarray = np.asarray(image.resize((84,84)), dtype='uint8')
            train_data[train_n] = img_ndarray
            train_label[train_n] = idx
            train_n += 1
assert(idx == 351-1 and train_n == 448695)
train_label[:5], train_label[1301], train_label[-5:], train_data[-1][-1][-1]
```

```python
val_data = np.empty((124261,84,84,3), dtype='uint8') # 124261 is the total number of samples in validation set.
val_label = np.empty(124261, dtype='uint16')
val_n = 0
with open(val_clsnames, "r") as f:
    for idx, line in enumerate(f):
        line = line.strip('\n')
        line = line.strip('\r')
        line = line.strip()
        print(line)
        all_classnames.append(line)
        for file in os.listdir(os.path.join(VAL_TIERED_IMAGE_DATASET_DIR, line)):
            image = Image.open(os.path.join(VAL_TIERED_IMAGE_DATASET_DIR, line, file)).convert('RGB')
            img_ndarray = np.asarray(image.resize((84,84)), dtype='uint8')
            val_data[val_n] = img_ndarray
            val_label[val_n] = idx + 351
            val_n += 1
assert(idx == 97-1 and val_n == 124261)
val_label[:5], val_label[1301], val_label[-5:], val_data[-1][-1][-1]
```

```python
test_data = np.empty((206209,84,84,3), dtype='uint8') # 206209 is the total number of samples in testing set.
test_label = np.empty(206209, dtype='uint16')
test_n = 0
with open(test_clsnames, "r") as f:
    for idx, line in enumerate(f):
        line = line.strip('\n')
        line = line.strip('\r')
        line = line.strip()
        print(line)
        all_classnames.append(line)
        for file in os.listdir(os.path.join(TEST_TIERED_IMAGE_DATASET_DIR, line)):
            image = Image.open(os.path.join(TEST_TIERED_IMAGE_DATASET_DIR, line, file)).convert('RGB')
            img_ndarray = np.asarray(image.resize((84,84)), dtype='uint8')
            test_data[test_n] = img_ndarray
            test_label[test_n] = idx + 351 + 97
            test_n += 1
assert(idx == 160-1 and test_n == 206209)
test_label[:5], test_label[1301], test_label[-5:], test_data[-1][-1][-1]
```

```python
all_classnames = np.array(all_classnames)
print(all_classnames.shape)
```

```python
_TIEREDIMAGENET_DATASET_DIR = '/data/FSLDatasets/tieredImageNet'
np.save(os.path.join(_TIEREDIMAGENET_DATASET_DIR, 'train_data.npy'), train_data)
np.save(os.path.join(_TIEREDIMAGENET_DATASET_DIR, 'train_label.npy'), train_label)
np.save(os.path.join(_TIEREDIMAGENET_DATASET_DIR, 'val_data.npy'), val_data)
np.save(os.path.join(_TIEREDIMAGENET_DATASET_DIR, 'val_label.npy'), val_label)
np.save(os.path.join(_TIEREDIMAGENET_DATASET_DIR, 'test_data.npy'), test_data)
np.save(os.path.join(_TIEREDIMAGENET_DATASET_DIR, 'test_label.npy'), test_label)
```
